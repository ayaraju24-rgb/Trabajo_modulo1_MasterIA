### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Promedio

esperanzas_de_vida= [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
años = 0

for vida in esperanzas_de_vida:
    años += vida

promedio = años / len(esperanzas_de_vida)

print("El promedio de la esperanza de vida es: {:.1f}".format(promedio), "años")

# Máximo
max = esperanzas_de_vida[0]  # empezamos suponiendo que el primero es el mayor
for valor in esperanzas_de_vida:
    if valor > max:
        max = valor

print(f"La edad máxima es: {max} años")

# Mínimo
min = esperanzas_de_vida[0]  # empezamos suponiendo que el primero es el menor
for valor in esperanzas_de_vida:
    if valor < min:
        min = valor

print(f"La edad mínima es: {min} años")
````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
print("         CLASIFICACIÓN ESPERANZA DE VIDA")

def clasificar_esperanza_vida(edad):
    if edad >= 75:
        return "Alta"
    elif edad >= 60:
        return "Media"
    else:
        return "Baja"


vida = []

while True:
    print("------------------------------------------")
    print("1. Registrar")
    print("2. Salir")

    op = input("Opción: ")

    if op == "1":
        try:
            edad = float(input("Ingrese los años de esperanza de vida: "))

            if edad < 0:
                print("Debe ser un número mayor o igual a 0")
                continue

            vida.append(edad)
            print(f"Clasificación: {clasificar_esperanza_vida(edad)}")

        except ValueError:
            print("Edad inválida, ingrese solo números")

    elif op == "2":
        break

    else:
        print("Opción inválida, digite 1 o 2")


if vida:
    print("\nEdades registradas:")

    for edad in vida:
        tipo = clasificar_esperanza_vida(edad)
        print(f"La edad registrada es {edad:.1f} años: Esperanza de vida {tipo.upper()}")
````

**Responde aquí** (doble clic para editar):

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

Primero, para preservar el tamaño del conjunto de datos y así no perder información relevante que halla en alguna otra columna. Por otro lado, reemplazando con el valor de la media no se altera el promedio general, por si se requiere en una operación futura.

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Esperanza de vida promedio por región
esperanza_vida_por_region = salud_copia.groupby("region")["esperanza_vida"].mean().round(1).to_frame()
print("ESPERANZA DE VIDA PROMEDIO POR REGIÓN (AÑOS)\n")
print(esperanza_vida_por_region)
print()

# Gasto en salud promedio por nivel de ingresos
gasto_salud_por_ingresos = salud_copia.groupby("nivel_ingresos")["gasto_salud_usd"].mean().round(1).to_frame()
print("GASTO EN SALUD PROMEDIO POR NIVEL DE INGRESOS\n")
print(gasto_salud_por_ingresos)
print()

# Número de paises por región
paises_por_region = salud_copia.groupby("region")["pais"].count().to_frame()
print("NÚMERO DE PAISES POR REGIÓN\n")
print(paises_por_region)
print()

vida_region = salud.groupby("region")["esperanza_vida"].mean()

# Región con la mayor esperanza de vida
print(f"{vida_region.idxmax()} es la región con la mayor esperanza de vida, con un promedio de: {vida_region.max():.1f} años")
print()

# Región con la menor esperanza de vida
print(f"{vida_region.idxmin()} es la región con la menor esperanza de vida, con un promedio de: {vida_region.min():.1f} años")
````

### 2.6 — Nueva columna calculada

🔧 **Ejercicio 2.6**
1. Crea una nueva columna llamada `categoria_vida` aplicando la función `clasificar_esperanza_vida` que creaste en la Parte 1 a la columna `esperanza_vida`
2. ¿Cuántos países caen en cada categoría? (usa `value_counts()`)
3. ¿Qué porcentaje del total representa cada categoría?

````python
# Categoría esperanza_vida y función clasificar_esperanza_vida
def clasificar_esperanza_vida(edad):
    if edad >= 75:
        return "Alta"
    elif edad >= 60:
        return "Media"
    else:
        return "Baja"

salud_copia["categoria_vida"] = salud_copia["esperanza_vida"].apply(clasificar_esperanza_vida)
print(" CLASIFICACIÓN CATEGORÍA - ESPERANZA DE VIDA\n")
print(salud_copia[["pais", "esperanza_vida", "categoria_vida"]].head(8).round(1))
print()

# Cantidad de países según la categoría de esperanza de vida
paises_por_categoria = salud_copia["categoria_vida"].value_counts()
print("NÚMERO DE PAÍSES SEGÚN LA CATEGORÍA DE ESPERANZA DE VIDA\n")
print(paises_por_categoria) 
print()

# Porcentaje de países por categoría de esperanza de vida
total_paises = salud_copia.shape[0]
print("PORCENTAJE DE PAÍSES POR CATEGORÍA DE ESPERANZA DE VIDA\n")
for categoria, cantidad in paises_por_categoria.items():
    porcentaje = (cantidad / total_paises) * 100
    print(f"{categoria}: {porcentaje:.1f}% ({cantidad} países)")
````
